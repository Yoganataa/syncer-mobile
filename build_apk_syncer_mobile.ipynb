# Build APK React Native / Expo di Google Colab
Gunakan notebook ini untuk mem-build project React Native / Expo (seperti **syncer-mobile**) langsung dari Google Colab.

Pastikan kamu sudah mengupload project kamu ke Google Drive dalam bentuk zip (misal: `syncer-mobile.zip`).

### 1. Mount Google Drive
Jalankan cell ini untuk menghubungkan Colab dengan Google Drive kamu.

```
from google.colab import drive
drive.mount('/content/drive')
```

### 2. Install Java 17
React Native modern membutuhkan Java 17.

```
!apt-get update -qq
!apt-get install -y openjdk-17-jdk-headless -qq > /dev/null
!update-alternatives --set java /usr/lib/jvm/java-17-openjdk-amd64/bin/java

import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-17-openjdk-amd64"
```

### 3. Install Node.js (Versi 20)
Dibutuhkan untuk menjalankan perintah npm / expo.

```
!curl -fsSL https://deb.nodesource.com/setup_20.x | bash - > /dev/null
!apt-get install -y nodejs -qq > /dev/null
!node -v
!npm -v
```

### 4. Install Android SDK

```
!wget -q https://dl.google.com/android/repository/commandlinetools-linux-10406996_latest.zip
!unzip -q commandlinetools-linux-10406996_latest.zip -d android-sdk
!mkdir -p android-sdk/cmdline-tools/latest
!mv android-sdk/cmdline-tools/bin android-sdk/cmdline-tools/lib android-sdk/cmdline-tools/source.properties android-sdk/cmdline-tools/latest/ 2>/dev/null || true

os.environ["ANDROID_HOME"] = "/content/android-sdk"
os.environ["PATH"] += ":/content/android-sdk/cmdline-tools/latest/bin:/content/android-sdk/platform-tools"

!yes | sdkmanager --licenses > /dev/null
!sdkmanager "platform-tools" "platforms;android-34" "build-tools;34.0.0" > /dev/null
```

### 5. Ekstrak Project Kamu
Ubah `/content/drive/MyDrive/syncer-mobile.zip` sesuai dengan lokasi file ZIP project kamu di Google Drive.

```
!unzip -q /content/drive/MyDrive/syncer-mobile.zip -d /content/syncer-mobile
```

### 6. Install Dependencies

```
%cd /content/syncer-mobile
!npm install
```

### 7. Build APK
Proses ini akan memakan waktu beberapa menit.

```
%cd /content/syncer-mobile/android
!chmod +x gradlew
!./gradlew assembleRelease
```

### 8. Download APK
Setelah build selesai, APK akan di-download otomatis.

```
from google.colab import files
files.download('/content/syncer-mobile/android/app/build/outputs/apk/release/app-release.apk')
```
